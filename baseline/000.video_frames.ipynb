# Frames

```python
from vs.video_file import VideoFilesProcessor
from vs.utils import find_all_files_with_pattern

BASE_DATA_DIR = "data"

video_processor = VideoFilesProcessor(
    base_dir=pathlib.Path(BASE_DATA_DIR), frame_dir=pathlib.Path("data/images")
)
import os
from tqdm import tqdm

video_files = find_all_files_with_pattern("*.MOV", ROOT_PATH / "data/video")
image_folder = ROOT_PATH / "data/images"
```

```python
video_files
```

```python
payloads = []

for v in video_files:
    payload = video_processor.process_single_video_file(v)
    payloads.append(payload)
```

# Storage

```python
from vs._local.storage import LocalVideoInfoStorage

video_storage = LocalVideoInfoStorage(payloads)

FILE_NAME = "data/video_metadata.json"
video_storage.dump_to_file(FILE_NAME)
```
